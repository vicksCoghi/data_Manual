```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
empresas = ['Apple','Google ' ,'Amazon',
            'Microsoft' ,'Tencent',"McDonald's",
            'Visa','Facebook','Alibaba',
            'Louis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646,
           611.460, 214.023, 191.032,
           191.032, 186.421, 169.966,
           124.273]
```

```python
plt.plot(empresas,bilhoes)
```

```python
plt.plot(empresas,bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.plot(empresas,bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilh√µes')
```

```python
#Redimensionando
plt.rcParams['figure.figsize'] = (15,8)
```

```python
plt.plot(empresas,bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilh√µes')
plt.show()
```

```python
plt.plot(empresas,bilhoes)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh√µes', fontsize = 20)
plt.show()
```

```python
#Mudando a cor e legendando
plt.plot(empresas,bilhoes, color = 'darkviolet')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh√µes', fontsize = 20)
plt.legend(['Bilh√µes de d√≥lares'],loc = 'right',fontsize=15)
plt.show()
```

```python
#aumentar a espessura da linha lw
plt.plot(empresas,bilhoes, color = 'darkviolet',lw=7)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh√µes', fontsize = 20)
plt.legend(['Bilh√µes de d√≥lares'],loc = 'right',fontsize=15)
plt.show()
```

```python
plt.plot(empresas,bilhoes, color = 'darkviolet',lw=7,marker='o')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh√µes', fontsize = 20)
plt.legend(['Bilh√µes de d√≥lares'],loc = 'right',fontsize=15)
plt.show()
```

```python
#Adicionando marcador (marker)
plt.plot(empresas,bilhoes, color = 'darkviolet',marker='o')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh√µes', fontsize = 20)
plt.legend(['Bilh√µes de d√≥lares'],loc = 'right',fontsize=15)
plt.show()
```

```python
#Adicionando linhas de grade
plt.plot(empresas,bilhoes, color = 'darkviolet',marker='o')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh√µes', fontsize = 20)
plt.legend(['Bilh√µes de d√≥lares'],loc = 'right',fontsize=15)
plt.grid(True)
plt.show()
```

```python
#Mudando o tipo do gr√°fico e colocando legenda

plt.bar(empresas,bilhoes, color = 'darkviolet')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh√µes', fontsize = 20)
plt.legend(['Bilh√µes de d√≥lares'],loc = 'center right',fontsize=15)
plt.grid(True)
plt.show()
```

```python
#gr√°fico simples 
plt.bar(empresas,bilhoes)
```

```python
cores = ['darkred','red','tomato','coral','salmon','deeppink','pink','indigo','darkviolet','violet']
```

```python
plt.bar(empresas,bilhoes,color = cores, edgecolor= 'white')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.ylabel('Bilh√µes de d√≥lares',fontsize=20)
plt.xlabel('Empresas',fontsize=20)
plt.grid()
```

```python
plt.style.available
```

```python
plt.style.use('dark_background')
plt.bar(empresas,bilhoes,color=cores,edgecolor='gray')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.ylabel('Bilh√µes de d√≥lares',fontsize=20)
plt.xlabel('Empresas',fontsize=20)
```

```python
#barra horizontal
plt.style.use('dark_background')
plt.barh(empresas,bilhoes,color=cores,edgecolor='white')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.ylabel('Bilh√µes de d√≥lares',fontsize=20)
plt.xlabel('Empresas',fontsize=20)
```
